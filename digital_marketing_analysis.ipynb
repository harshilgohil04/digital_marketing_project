```python
discrete_variables = ["Age", "WebsiteVisits", "SocialShares", "EmailOpens", "EmailClicks", "PreviousPurchases"]
continuous_variables = ["Income", "AdSpend", "ClickThroughRate", "ConversionRate", "PagesPerVisit", "TimeOnSite", "LoyaltyPoints"]
categorical_variables = ["Gender", "CampaignType", "AdvertisingPlatform", "AdvertisingTool", "Conversion"]
```

```python
df.describe().T
```

```python
# Set up the plotting environment with larger graphs and 2 graphs per row
plt.figure(figsize=(20, 20))

# Loop through discrete variables and create histograms with mean, median, and mode
for i, feature in enumerate(discrete_variables):
    plt.subplot(3, 2, i + 1)
    sns.histplot(df[feature], kde=True, bins=30, color='orange')
    
    # Calculate and plot mean, median, and mode
    mean_val = df[feature].mean()
    median_val = df[feature].median()
    mode_val = df[feature].mode()[0]
    
    plt.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}')
    plt.axvline(median_val, color='green', linestyle='--', label=f'Median: {median_val:.2f}')
    plt.axvline(mode_val, color='blue', linestyle='--', label=f'Mode: {mode_val:.2f}')
    
    plt.title(f'Distribution of {feature}')
    plt.legend()
    
    #add statistics on the plot
    plt.text(mean_val, plt.ylim()[1]*0.9, f'Mean: {mean_val:.2f}', color='red', ha='center',fontweight='bold',fontsize=14)
    plt.text(median_val, plt.ylim()[1]*0.7, f'Median: {median_val:.2f}', color='green', ha='center',fontweight='bold',fontsize=14)
    plt.text(mode_val, plt.ylim()[1]*0.5, f'Mode: {mode_val:.2f}', color='blue', ha='center',fontweight='bold',fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# Set up the plotting environment with larger graphs and 2 graphs per row
plt.figure(figsize=(20, 30))

# Loop through discrete variables and create histograms with mean, median, and mode
for i, feature in enumerate(continuous_variables):
    plt.subplot(4, 2, i + 1)
    sns.histplot(df[feature], kde=True, bins=30, color='orange')
    
    # Calculate and plot mean, median, and mode
    mean_val = df[feature].mean()
    median_val = df[feature].median()
    mode_val = df[feature].mode()[0]
    
    plt.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}')
    plt.axvline(median_val, color='green', linestyle='--', label=f'Median: {median_val:.2f}')
    plt.axvline(mode_val, color='blue', linestyle='--', label=f'Mode: {mode_val:.2f}')
    
    plt.title(f'Distribution of {feature}')
    plt.legend()
    
    #add statistics on the plot
    plt.text(mean_val, plt.ylim()[1]*0.9, f'Mean: {mean_val:.2f}', color='red', ha='center',fontweight='bold',fontsize=14)
    plt.text(median_val, plt.ylim()[1]*0.7, f'Median: {median_val:.2f}', color='green', ha='center',fontweight='bold',fontsize=14)
    plt.text(mode_val, plt.ylim()[1]*0.5, f'Mode: {mode_val:.2f}', color='blue', ha='center',fontweight='bold',fontsize=14)

plt.tight_layout()
plt.show()
```

The Low Ad Spend category has a significantly higher ROAS (52.50) compared to the other bins. This suggests that when spending is low, the ads are more cost-effective, yielding a higher return per dollar spent.

**Diminishing Returns with Increased Spend:**

As the ad spend increases from Medium to Very High, the ROAS consistently decreases. This could be due to several factors:
Market Saturation: Higher spending might not be reaching new audiences effectively, resulting in diminishing returns.
Ad Fatigue: As spend increases, the same audience might see the ads multiple times, reducing their effectiveness.
Poor Targeting: Increased spend might be allocated to less relevant audiences, reducing conversion rates.

**Inefficient Allocation in High Spends:**
